```python
import torch.distributions as dist
import torch
```

```python
params = {"A": (3, 10),
         "B": (250, 400),
         "C": (153, 450),
         }

colors = ["g", "b", "k"]
lines = ["--", "-.", ":", "-"]
(Hitters.Hits/Hitters.AtBat).hist(bins=50, alpha=0.2)

x_lin = torch.linspace(0.0, 0.5, 100)
for i, (name, d) in enumerate(params.items()):
    plt.plot(x_lin, dist.Beta(*d).log_prob(x_lin).exp(), 
             label=f"{name}: Beta{d}", lw=4, ls=lines[i], c=colors[i])
plt.legend(title="Prior")
plt.grid(False)
plt.xlabel("Batting Average")
plt.savefig("baseball-b.pdf")
```

```python
import numpy as np
from scipy import stats

# Your data
data = (Hitters.Hits / Hitters.AtBat).values

# Define the distribution you want to fit (for example, a beta distribution)
distribution = stats.beta



# Specify bounds for the parameters
bounds = ((1, 300), (1, 300))

# Fit the distribution with bounds
res = stats.fit(distribution, data, bounds=bounds)
```

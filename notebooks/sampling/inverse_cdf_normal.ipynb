# Inverse CDF Sampling

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# Retina display
%config InlineBackend.figure_format = 'retina'
import warnings
warnings.filterwarnings('ignore')
```

```python
from tueplots import bundles
plt.rcParams.update(bundles.beamer_moml())


# Also add despine to the bundle using rcParams
plt.rcParams['axes.spines.right'] = False
plt.rcParams['axes.spines.top'] = False

# Increase font size to match Beamer template
plt.rcParams['font.size'] = 16
# Make background transparent
plt.rcParams['figure.facecolor'] = 'none'
```

```python
mu = 0.0
sigma = 1.0
num_samples = [25, 100, 1000]
```

```python
u1 = torch.rand(num_samples[0], generator=torch.Generator().manual_seed(1))
u2 = torch.rand(num_samples[1], generator=torch.Generator().manual_seed(2))
u3 = torch.rand(num_samples[2], generator=torch.Generator().manual_seed(3))
x_samples1 = mu + sigma * torch.erfinv(2*u1 - 1) * (2 ** 0.5)
x_samples2 = mu + sigma * torch.erfinv(2*u2 - 1) * (2 ** 0.5)
x_samples3 = mu + sigma * torch.erfinv(2*u3 - 1) * (2 ** 0.5)
```

```python
fig, ax = plt.subplots(1, 2)
ax[0].axhline(1, linewidth=0.5)
ax[0].hist(u1.numpy(), bins=10, density=True)
ax[0].set_title('Samples from Uniform Distribution')
ax[0].set_xlabel('Value')
ax[0].set_ylabel('Density')
ax[1].hist(x_samples1.numpy(), bins=10, density=True)
sns.kdeplot(x_samples1.numpy())
ax[1].set_title('Approximated Normal Distribution')
ax[1].set_xlabel('Value')
ax[1].set_ylabel('Density')
# plt.savefig('figures/sampling/inverse_cdf_normal1.pdf', transparent=True, bbox_inches='tight', dpi=300)
```

```python
fig, ax = plt.subplots(1, 2)
ax[0].axhline(1, linewidth=0.5)
ax[0].hist(u2.numpy(), bins=10, density=True)
ax[0].set_title('Samples from Uniform Distribution')
ax[0].set_xlabel('Value')
ax[0].set_ylabel('Density')
ax[1].hist(x_samples2.numpy(), bins=10, density=True)
sns.kdeplot(x_samples2.numpy())
ax[1].set_title('Approximated Normal Distribution')
ax[1].set_xlabel('Value')
ax[1].set_ylabel('Density')
# plt.savefig('figures/sampling/inverse_cdf_normal2.pdf', transparent=True, bbox_inches='tight', dpi=300)
```

```python
fig, ax = plt.subplots(1, 2)
ax[0].axhline(1, linewidth=0.5)
ax[0].hist(u3.numpy(), bins=10, density=True)
ax[0].set_title('Samples from Uniform Distribution')
ax[0].set_xlabel('Value')
ax[0].set_ylabel('Density')
ax[1].hist(x_samples3.numpy(), bins=10, density=True)
sns.kdeplot(x_samples3.numpy())
ax[1].set_title('Approximated Normal Distribution')
ax[1].set_xlabel('Value')
ax[1].set_ylabel('Density')
# plt.savefig('figures/sampling/inverse_cdf_normal3.pdf', transparent=True, bbox_inches='tight', dpi=300)
```
